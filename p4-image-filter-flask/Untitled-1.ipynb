```python
import requests
import json
from PIL import Image
import io

#  Get usage info from the API
response = requests.post("http://127.0.0.1:5000/")
response.raise_for_status()
info = json.loads(response.content.decode("utf-8"))
print("Available filters:", info["filters_available"])

# Send image with filter
file = { "image": open(r"D:\rrcc\web development3\web-apis-with-python\p4-image-filter-flask\sample.jpg", "rb") }
URL = "http://127.0.0.1:5000"
filter = "detail"
endpoint = f"{URL}/{filter}"

response = requests.post(endpoint, files=file)
response.raise_for_status()

# Convert response to image
image = Image.open(io.BytesIO(response.content))
image.save("response.jpg", "JPEG")
image  # opens the image viewer
```
